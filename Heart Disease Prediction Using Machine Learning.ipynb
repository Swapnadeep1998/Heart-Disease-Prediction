```python
from sklearn.model_selection import train_test_split
```

```python
#data = pd.get_dummies(data, columns = ['sex','cp','fbs','restecg','exang','slope','ca','thal'])
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler =  StandardScaler()
columns_to_scale = ['age','trestbps','chol','thalach','oldpeak']
data[columns_to_scale] = std_scaler.fit_transform(data[columns_to_scale])
```

```python
y = data['target']
X = data.drop(['target'],1)
```

```python
X
```

```python
y
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
y_train
```

# KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
y_pred_knn = knn.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_knn))
print(confusion_matrix(y_test,y_pred_knn))
```

```python
e_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred_knn = knn.predict(X_test)
    e_rate.append(np.mean(y_pred_knn != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),e_rate,marker = 'o',markerfacecolor = 'red',markersize = 7)
```

```python
knn = KNeighborsClassifier(n_neighbors=27)
knn.fit(X_train,y_train)
y_pred_knn = knn.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_knn))
print(confusion_matrix(y_test,y_pred_knn))
```

# Decision Tree Classifier Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred_dtc = dtc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_dtc))
print(confusion_matrix(y_test,y_pred_dtc))
```

# Random Forest Classifier Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
rdfc = RandomForestClassifier(n_estimators=800)
rdfc.fit(X_train,y_train)
y_pred_rdfc = rdfc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_rdfc))
print(confusion_matrix(y_test,y_pred_rdfc))
```
